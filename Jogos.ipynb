# Jogo da Adivinhação!

```python
import random
def jogar_adivinhacao():
  print('*********************************')
  print('Bem vindo ao jogo de Adivinhação!')
  print('*********************************')

  numero_secreto = random.randrange(1, 101)
  total_de_tentativas = 0
  pontos = 1000

  print('\nQual é o nível de dificuldade?')
  print('\n(1) Fácil (2) Médio (3) Difícil\n')


  nivel = int(input('Definir o nível: '))

  if nivel == 1:
    total_de_tentativas = 20
  elif nivel == 2:
    total_de_tentativas = 10
  else:
    total_de_tentativas = 5


  for rodada in range(1, total_de_tentativas+1):
    print('~~~~~~~~~~~~~~~~')
    print(f'Tentativa {rodada} de {total_de_tentativas}')
    print('~~~~~~~~~~~~~~~~')
    print('\n')
    chute_str = input('Digite um número entre 1 e 100: ')
    chute = int(chute_str)

    if (chute < 1 or chute > 100):
      print('\n')
      print('Você deve digitar um número entre 1 e 100!')
      print('\n')
      continue

    acertou = chute == numero_secreto
    errou_maior = chute > numero_secreto
    errou_menor = chute < numero_secreto

    print('\n')

    print(f'Você digitou {chute} \n')

    if acertou:
      print(f'Você acertou e fez {pontos} pontos!')
      break
    else:
      if errou_maior:
        print('Você errou! O seu chute foi maior doque o número secreto.')
        print('\n')
      elif errou_menor:
        print('Você errou! O seu chute foi menor doque o número secreto.')
        print('\n')
      pontos_perdidos = abs(numero_secreto - chute)
      pontos = pontos - pontos_perdidos
    
  print('\n')

  print(f'Número secreto = {numero_secreto}\n\nFim do jogo!')

if (__name__ == '__main__'):
  jogar_adivinhacao()
```

# Jogo da Forca!

```python
import random


def jogar_forca():

  imprime_mensagem_de_abertura()
  palavra_secreta = carrega_palavra_secreta()
  letras_acertadas = inicializa_letras_acertadas(palavra_secreta)
  print('\n',letras_acertadas,'\n')

  enforcou = False
  acertou = False
  erros = 0
  
  while not enforcou and not acertou:
    
    chute = pede_chute()

    if chute in palavra_secreta:
      marca_chute_correto(chute, letras_acertadas, palavra_secreta)      
    else:
      erros += 1

    enforcou = erros == 7
    tentativas = 7 - erros

    print('\n',letras_acertadas)
    letras_faltando = str(letras_acertadas.count('_'))
    acertou = letras_faltando == '0'
    
    if letras_faltando == '0':
      ganhador()
    elif erros == 7:
      perderdor()
    else:
      errou(letras_faltando, tentativas, erros)
  
  print(f'\nPalavra secreta = {palavra_secreta}\n\nFim do jogo!')


def imprime_mensagem_de_abertura():
  print('*********************************')
  print('***Bem vindo ao jogo da Forca!***')
  print('*********************************')

def carrega_palavra_secreta():
  arquivo = open('arquivo.txt', 'r')
  palavras = []
  for linha in arquivo:
    linha = linha.strip()
    palavras.append(linha)
  arquivo.close()
  palavra_secreta = palavras[random.randrange(0,len(palavras))].upper()
  return palavra_secreta

def inicializa_letras_acertadas(palavra):
  return ['_' for letra in palavra]

def pede_chute():
  chute = input('Qual a letra? ')
  chute = chute.strip().upper()
  return chute

def marca_chute_correto(chute, letras_acertadas, palavra_secreta):
  index = 0
  for letra in palavra_secreta:
    if chute == letra:
      letras_acertadas[index] = letra
    index += 1

def ganhador():
  print('\nParabéns você ganhou!\n')
  print("       ___________      ")
  print("      '._==_==_=_.'     ")
  print("      .-\\:      /-.    ")
  print("     | (|:.     |) |    ")
  print("      '-|:.     |-'     ")
  print("        \\::.    /      ")
  print("         '::. .'        ")
  print("           ) (          ")
  print("         _.' '._        ")
  print("        '-------'       ")

def perderdor():
  print('\nQue pena, você se enforcou!\n')
  print("    _______________         ")
  print("   /               \       ")
  print("  /                 \      ")
  print("//                   \/\  ")
  print("\|   XXXX     XXXX   | /   ")
  print(" |   XXXX     XXXX   |/     ")
  print(" |   XXX       XXX   |      ")
  print(" |                   |      ")
  print(" \__      XXX      __/     ")
  print("   |\     XXX     /|       ")
  print("   | |           | |        ")
  print("   | I I I I I I I |        ")
  print("   |  I I I I I I  |        ")
  print("   \_             _/       ")
  print("     \_         _/         ")
  print("       \_______/           ")

def errou(letras_faltando, tentativas, erros):
  print(f'\nAinda faltam acertar {letras_faltando} letras e você ainda possui {tentativas} tentativas!\n')
  print("  _______     ")
  print(" |/      |    ")
  if erros == 1:
    print(" |      (_)   ")
    print(" |            ")
    print(" |            ")
    print(" |            ")
  elif erros == 2:
    print(" |      (_)   ")
    print(" |      \     ")
    print(" |            ")
    print(" |            ")
  elif erros == 3:
    print(" |      (_)   ")
    print(" |      \|    ")
    print(" |            ")
    print(" |            ") 
  elif erros == 4:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |            ")
    print(" |            ")
  elif erros == 5:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |            ")
  elif erros == 6:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |      /     ")
  elif erros == 7:
    print(" |      (_)   ")
    print(" |      \|/   ")
    print(" |       |    ")
    print(" |      / \   ")
  print(" |            ")
  print("_|___         ")
  print()

if (__name__ == '__main__'):
  jogar_forca()
```
